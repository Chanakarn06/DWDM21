**เริ่มทำจ้าาา**

````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน')  
#ทำการเชื่อมตารางทั้ง 2 ตาราง คือตาราง flights และตาราง airports 
````

````python
merged_table
````

````python
merged_table.dropna() ##เช็คค่า missing
````

# 1)หารัฐที่มีจำนวนเครื่องบินขาออกมากที่สุดและน้อยที่สุด

````python
# เช็คจำนวนเที่ยวบินของแต่ละรัฐ
merged_table.groupby('รัฐ').count() 
 
````

````python
 # รัฐที่มีสนามบินปลายทางมากสุด 
RWmax = merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) 
RWmax.iloc[:1,4:5]
````

````python
# รัฐที่มีสนามบินปลายทางน้อยสุด 
RWmin = merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) 
RWmin.iloc[-1:,4:5]
````

# 4) สายการบินอะไรมีความล่าช้ามากที่สุดและน้อยที่สุด

````python
# เช็คความล่าช้าขาออกที่มีค่ามากสุด
merged_table['ความล่าช้าขาออก'].max() 
````

````python
# เช็คความล่าช้าขาออกที่มีค่ามากสุด
merged_table['ความล่าช้าขาเข้า'].max() 
````

````python
#ความล่าช้าขาออกมากสุด
DLmax = merged_table.groupby('สายการบิน').max().sort_values(by=['ความล่าช้าขาออก'],ascending=False) 
DLmax.iloc[:1,4:5]
````

````python
#ความล่าช้าขาเช้ามากสุด
DLmax1= merged_table.groupby('สายการบิน').max().sort_values(by=['ความล่าช้าขาเข้า'],ascending=False) 
DLmax1.iloc[:1,5:6]
````

````python
# เช็คความล่าช้าขาออกที่มีค่าน้อยสุด
merged_table['ความล่าช้าขาออก'].min() 
````

````python
# เช็คความล่าช้าขาเข้าที่มีค่าน้อยสุด
merged_table['ความล่าช้าขาเข้า'].min() 
````

````python
#ความล่าช้าขาออกน้อยสุด
DLmin = merged_table.groupby('สายการบิน').min().sort_values(by=['ความล่าช้าขาออก'],ascending=False) 
DLmin.iloc[-1:,4:5]
````

````python
#ความล่าช้าขาออกน้อยสุด
DLmin1 = merged_table.groupby('สายการบิน').min().sort_values(by=['ความล่าช้าขาเข้า'],ascending=False) 
DLmin1.iloc[-1:,5:6]
````
